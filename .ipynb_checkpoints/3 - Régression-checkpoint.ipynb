```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore')

from sklearn.model_selection import train_test_split
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(url, delimiter=';')
df = data.copy()
df.shape
```

```python
seed = 0
trainset, testset = train_test_split(df, test_size=0.25, random_state=seed)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x='quality', data=trainset)
plt.title('RÃ©partition de la variable quality dans le trainset')
plt.subplot(1,2,2)
sns.countplot(x='quality', data=testset)
plt.title('RÃ©partition de la variable quality dans le testset')
```

On voit que les classes sont de tailles trÃ¨s dÃ©sÃ©quilibrÃ©es, on peut se demander si ces dispartÃ©s ont un impact sur la qualitÃ© de la rÃ©gresion


# Preprocessing + Feature selection

```python
from collections import Counter

def preprocessing(df, resampler=None):
    
    df_unique = df.drop_duplicates()
    
    X = df_unique.drop('quality', axis=1)
    y = df_unique.quality
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=seed)
    if resampler!=None:
        X_train,y_train = resampler.fit_resample(X_train,y_train)
    
    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = preprocessing(df, resampler=None)
print('Original dataset shape %s' % Counter(y_train))
plt.scatter(X_train['alcohol'],X_train['volatile acidity'], marker='o', c=y_train, s=25, edgecolor='k', cmap=plt.cm.coolwarm)
```

On va utiliser les mÃ©thodes SMOTE et SMOTETomek construire des nuages de donnÃ©es plus Ã©quilibrÃ©s. L'idÃ©e est de crÃ©er des points supplÃ©mentaires dans les classes sous-reprÃ©sentÃ©es Ã  partir de combinaisons convexes de points de ces classes.

### Nuage de points SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=seed, k_neighbors=4)
X_train, y_train, X_test, y_test = preprocessing(df, resampler=smote)
print('Resampled dataset shape %s' % Counter(y_train))
plt.scatter(X_train['alcohol'],X_train['volatile acidity'], marker='o', c=y_train, s=25, edgecolor='k', cmap=plt.cm.coolwarm)
```

### Nuage de points SMOTETomek

```python
from imblearn.combine import SMOTETomek
smt = SMOTETomek(random_state=seed)
X_train, y_train, X_test, y_test = preprocessing(df, resampler=smt)
print('Resampled dataset shape %s' % Counter(y_train))
plt.scatter(X_train['alcohol'],X_train['volatile acidity'], marker='o', c=y_train, s=25, edgecolor='k', cmap=plt.cm.coolwarm)
```

# ProcÃ©dure d'Ã©valuation

DÃ©terminons maintenant si ces manipulations amÃ©liorent ou non les rÃ©sultats

```python
from sklearn.model_selection import learning_curve
from sklearn.metrics import mean_squared_error

def evaluation_mse(model, graph = True):

    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f'{model[-1].__class__.__name__ :<25} {round(mse,3)} MSE score.')
    
    if graph:
        fig, ax = plt.subplots(1, 3, figsize = (16,4))
        fig.subplots_adjust(top=0.85)
        
    #     ax[0].set_title('Matrice de confusion')
    #     cm = pd.DataFrame(confusion_matrix(y_test, y_pred))
    #     sns.heatmap(cm, annot=True, cmap='Blues', fmt="d", ax=ax[0])
    #     ax[0].set_xlabel('Predicted label')
    #     ax[0].set_ylabel('True label')
        
        ax[0].set_title('y_test vs y_pred')
        ax[0].scatter(x = y_test, y = y_pred)
        ax[0].set_xlabel('Predicted label')
        ax[0].set_ylabel('True label')
        
        ax[1].set_title('Residual Plot of test samples')
        sns.distplot((y_test-y_pred), hist = False, ax = ax[1])
        ax[1].set_xlabel('y_pred')
        
        ax[2].set_title('Learning curve')
        N, train_score, val_score = learning_curve(model, X_train, y_train,
                                                cv = 4,
                                                scoring = 'neg_mean_absolute_error',
                                                train_sizes = np.linspace(0.1, 1, 10))
        ax[2].plot(N, train_score.mean(axis=1), label='train score')
        ax[2].plot(N, val_score.mean(axis=1), label='validation score')
        
        plt.suptitle(model[-1].__class__.__name__, fontsize=14)
        plt.legend();
```

```python
from sklearn.metrics import r2_score

def evaluation_r2(model, graph = True):

    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f'{model[-1].__class__.__name__ :<25} {round(r2,3)} R2 score.')
    
    if graph:
        fig, ax = plt.subplots(1, 3, figsize = (16,4))
        fig.subplots_adjust(top=0.85)
        
    #     ax[0].set_title('Matrice de confusion')
    #     cm = pd.DataFrame(confusion_matrix(y_test, y_pred))
    #     sns.heatmap(cm, annot=True, cmap='Blues', fmt="d", ax=ax[0])
    #     ax[0].set_xlabel('Predicted label')
    #     ax[0].set_ylabel('True label')
        
        ax[0].set_title('y_test vs y_pred')
        ax[0].scatter(x = y_test, y = y_pred)
        ax[0].set_xlabel('Predicted label')
        ax[0].set_ylabel('True label')
        
        ax[1].set_title('Residual Plot of test samples')
        sns.distplot((y_test-y_pred), hist = False, ax = ax[1])
        ax[1].set_xlabel('y_pred')
        
        ax[2].set_title('Learning curve')
        N, train_score, val_score = learning_curve(model, X_train, y_train,
                                                cv = 4,
                                                scoring = 'r2_score',
                                                train_sizes = np.linspace(0.1, 1, 10))
        ax[2].plot(N, train_score.mean(axis=1), label='train score')
        ax[2].plot(N, val_score.mean(axis=1), label='validation score')
        
        plt.suptitle(model[-1].__class__.__name__, fontsize=14)
        plt.legend();
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
# models
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
SVR = make_pipeline(StandardScaler(), SVR())
OLS = make_pipeline(StandardScaler(), LinearRegression())
PCR = make_pipeline(StandardScaler(), PCA(n_components=2), LinearRegression())
PLS = make_pipeline(StandardScaler(), PLSRegression(n_components=2))
KNN = make_pipeline(KNeighborsRegressor())
RandomForest = make_pipeline(RandomForestRegressor(random_state=0))
XGBoost = make_pipeline(StandardScaler(), XGBRegressor(verbosity=0, random_state=0))
```

```python
dict_of_models = {'SVM': SVR,
                  'OLS':OLS,
                  'PCR':PCR,
                  'PLS':PLS,
                  'KNN': KNN,
                  'RandomForest': RandomForest,
                  'XGBoost': XGBoost
                  }
```

```python
#no preprocessing
X_train, y_train, X_test, y_test = preprocessing(df, resampler=None)
for model in dict_of_models.values():
    evaluation_mse(model, graph=False)
    evaluation_r2(model, graph=False)
```

# ModÃ©lisation

On voit que les grandes diffÃ©rences de tailles entre les classes n'ont pas l'air d'alterer la qualitÃ© de la rÃ©gression. On va pour terminer en effectuer plusieurs.

```python
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

df_unique = df.drop_duplicates()
X = df_unique.drop('quality', axis=1)
y = df_unique.quality
y = y.values
y = y.reshape((len(y),1))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score

r2 = []
mse = []
rpd = []
for n_comp in range(1,10):
    pls = PLSRegression(n_components=n_comp)

    # Cross-validation
    y_cv = cross_val_predict(pls, X, y, cv=10)

    # Calculate scores
    r2.append(r2_score(y, y_cv))
    MSE = mean_squared_error(y, y_cv)
    mse.append(MSE)
    rpd.append(y.std()/np.sqrt(MSE))

fig, ax = plt.subplots(1, 3, figsize = (16,4))
fig.subplots_adjust(top=0.85)
ax[0].plot(range(1,10),r2,label = 'r2')
ax[1].plot(range(1,10),mse, color='r', label = 'mse')
ax[2].plot(range(1,10),rpd, color='g', label = 'rpd')
fig.legend()
```

On va retenir 7 composantes

## PLS

```python
pls = PLSRegression(n_components=7)
model = pls.fit(X_train,y_train)
y_pred = model.predict(X_test)

plt.scatter(y_test, y_pred, color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.plot()
print('score : ', model.score(X_test,y_test))
```

## Regression linÃ©aire

```python
model = LinearRegression().fit(X_train,y_train)
y_pred = model.predict(X_test)

plt.scatter(y_test, y_pred, color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.plot()
print('intercept : ',model.intercept_,'coefficients : ',model.coef_)
print('score : ', model.score(X_test,y_test))
```

```python
r2=[]
mse=[]
rpd=[]
for n in range(1,40):
    KNN=KNeighborsRegressor(n_neighbors=n)

    # Cross-validation
    y_cv = cross_val_predict(KNN, X, y, cv=10)

    # Calculate scores
    r2.append(r2_score(y, y_cv))
    MSE = mean_squared_error(y, y_cv)
    mse.append(MSE)
    rpd.append(y.std()/np.sqrt(MSE))

fig, ax = plt.subplots(1, 3, figsize = (16,4))
fig.subplots_adjust(top=0.85)
ax[0].plot(range(1,40),r2,label = 'r2')
ax[1].plot(range(1,40),mse, color='r', label = 'mse')
ax[2].plot(range(1,40),rpd, color='g', label = 'rpd')
plt.legend()
```
